```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ruleset
```

```python
data = pd.read_csv("people.txt", delimiter=' ')
```

```python
data
```

```python
rules = []
rules.append(ruleset.age_check)
rules.append(ruleset.age_check2)
rules.append(ruleset.status_check)
rules.append(ruleset.agegroup_check)
```

```python
violations = []
# map(function,iterations)  ##map returns iterable object
rule = list(map(lambda x: "rule"+str(x), range(1, len(rules)+1)))
for i in range(len(rules)):
    violation, desc = rules[i](data)
    print(f"{rule[i]}: {desc}\nviolations:{violation}")
    violations.append(violation)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.bar(rule, violations, 0.3, color='green')  # []
plt.ylabel("Number of violations")
plt.xlabel("rules")
plt.title("Plotting violation in rules")
plt.show()
```
