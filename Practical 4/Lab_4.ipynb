The dataset is bread basket

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
```

```python
grocery_df = pd.read_csv("bread basket.csv")
```

```python
grocery_df
```

The items are converted into list according to transaction

```python
df = grocery_df.groupby('Transaction')['Item'].apply(
    list).reset_index(name='Item List')
df
```

The data is encoded

```python
from mlxtend.preprocessing import TransactionEncoder
encoded = TransactionEncoder()
encoded_array = encoded .fit_transform(df['Item List'])
encoded_array
```

```python
arr_to_df = pd.DataFrame(encoded_array, columns=encoded.columns_)
```

Minimum support is taken as 3% and threshold as 50%

```python
frequent_itemsets = apriori(arr_to_df, min_support=0.03, use_colnames=True)
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(
    frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```
