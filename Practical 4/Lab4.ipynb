The dataset is groceries.csv

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
```

```python
grocery_df = pd.read_csv("groceries.csv")
```

```python
grocery_df
```

```python
grocery_df.shape
```

```python
items_list = grocery_df['Items'].str.split(',').values
```

```python
items_list.tolist()
```

```python
from mlxtend.preprocessing import TransactionEncoder
encoded = TransactionEncoder()
encoded_array = encoded .fit_transform(items_list)
encoded_array
```

```python
arr_to_df = pd.DataFrame(encoded_array, columns=encoded.columns_)
```

Minimum support is taken as 1% and threshold as 75%

```python
frequent_itemsets = apriori(arr_to_df, min_support=0.01, use_colnames=True)
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(
    frequent_itemsets, metric="confidence", min_threshold=0.75)
rules
```
